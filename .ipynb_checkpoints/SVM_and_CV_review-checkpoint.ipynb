```python
data1 = pd.read_csv('sample_dataset_1.csv').drop('Unnamed: 0',axis=1) # dummy data 1
#data1.drop('Unnamed: 0',axis=1,inplace=True)

X=data1.iloc[:,:2]
Y=data1.iloc[:,2]

plt.scatter(X.iloc[:,0], X.iloc[:,1], s=100, c=Y, cmap=plt.cm.get_cmap('jet', 2))
plt.axis('tight')
plt.show()

data1.head(4)
```

```python
# go with Linear Kernal for now

clf = svm.SVC(kernel='linear',C=100000)
clf.fit(X, Y)
```

### Boundary formation...

```python
# Hyperplane splitting with clear boundaries

w = clf.coef_[0]
h_slope = -w[0] / w[1]
h_intercept = -clf.intercept_[0] / w[1]

x1 = np.linspace(-4,4)
y1 = h_slope*x1 - h_intercept

print "Hyperplane: y=",h_slope,"* x -",-h_intercept
print ""

plt.plot(x1, y1, 'k-')
plt.scatter(X.iloc[:,0], X.iloc[:,1], s=100, c=Y, cmap=plt.cm.get_cmap('jet', 2));

plt.axis('tight')
plt.show()
```

```python
# Measure/visualize support vectors and margins...

b_down = clf.support_vectors_[0] # lower support vectors
b_up = clf.support_vectors_[-1] # upper support vectors

y_downward = h_slope * x1 + (b_down[1] - h_slope * b_down[0])
print "Downward margin line: y=", h_slope,"* x -", -(b_down[1] - h_slope * b_down[0])

y_upward = h_slope * x1 + (b_up[1] - h_slope * b_up[0])
print "Upward margin line: y=", h_slope,"* x +",(b_up[1] - h_slope * b_up[0])

margin = 2 / np.sqrt(np.sum(clf.coef_ ** 2))
print ""; print "Margin Width (Perpenidcular upper to lower line): ",margin

print ""
print "Support Vectors:";print clf.support_vectors_

plt.plot(x1, y1, 'k-')
plt.plot(x1, y_downward, 'k--')
plt.plot(x1, y_upward, 'k--')
plt.scatter(clf.support_vectors_[:, 0], clf.support_vectors_[:, 1], s=600, facecolors='none')
plt.scatter(X.iloc[:,0], X.iloc[:,1], s=60, c=Y, cmap=plt.cm.get_cmap('jet', 2));

plt.axis('tight')

plt.show()
```

### Tune SVM Hyperparameters *degree* & *C* (without GS/sklearn):

```python
# Cross validate to get best C hyperparemter...

def SVM_fit_C(data,Cs,deg,Graph=False, logl=True):

    OS_acc = []
    OS_acc_M=[]    

    SVM_fit = lambda a: svm.SVC(kernel='poly',C=a,degree=deg) 
    
    best_accuracy = -1
    best_C = Cs[0]
    
    for i in Cs:
        #print i
        for j in range(0,10):    #10-fold  (not really k-folds here, take note)
            dt = data_train.sample(frac=1)  # shuffle sample (a good idea here???).  vali set can be used more than once
            y_train = dt['y']
            X_train = dt.drop(['y'],axis=1)
            X_train0, X_valid0, y_train0, y_valid0 = train_test_split(X_train, 
                y_train, test_size = 0.33, random_state = 200)
            mod = SVM_fit(i)
            mod.fit(X_train0,y_train0)  
            OS_acc_ = 1.0*mod.score(X_valid0, y_valid0) #.max()
            OS_acc.append(OS_acc_)
    
        OS_acc_M_ = np.mean(OS_acc) # get mean of 10 validation runs
        OS_acc = []
        OS_acc_M.append(OS_acc_M_)
        if OS_acc_M_ > best_accuracy:
            best_accuracy = OS_acc_M_ 
            best_C = i
 
    if Graph==True:  # visualize better
        plt.title('Accuracy across Cs')
        if logl:
            plt.xlabel('ln(C)')
            l=np.log10(Cs)
            bl=np.log10(best_C)
        else:
            plt.xlabel('C')
            l=Cs
            bl=best_C
        plt.plot(l,OS_acc_M,'b')
        plt.legend(loc='upper right')
        plt.ylabel('Accuracy')
        plt.axvline(bl,color='r',linestyle='--')
        plt.ylim(-.1,1)

        plt.show()
    
    print 
    return best_C, best_accuracy

data_train.sort_index(inplace=True)
```

```python
# try each C with different Degrees

current_best_acc = 0

for i in range(1,5,1):
    #C_hyp = [math.exp(i) for i in np.linspace(-10,2*deg,200)]
    C_hyp = np.exp(np.linspace(-10,2*i,200))
    C_optimal,best_acc=SVM_fit_C(data_train,C_hyp,i,Graph=True,logl=True)
    if best_acc > current_best_acc:
        current_best_acc = best_acc
        best_degree = i
        best_C = C_optimal
    print('At Degree {1}, model is most accurate when C = {0} (Accuracy = {2})'.format(C_optimal,i,best_acc))

print "";""
print('Best Degree = {0} with Best C = {1} ...where accuracy is best (at {2})'.format(best_degree,best_C,current_best_acc))
print ""

```

```python
# report best hyperparameters from cross-valid


print "\nAccuracy is at its max (",current_best_acc,") when degree is",best_degree,"and C is", best_C


dt = data_train.sample(frac=1)  # is this fair (??)  and again (??)

y_train = dt['y']
X_train = dt.drop(['y'],axis=1)

y_test = data_test['y']
X_test = data_test.drop(['y'],axis=1)

X_train0, X_valid0, y_train0, y_valid0 = train_test_split(X_train, 
                                                          y_train, test_size = 0.33, random_state = 200)


print "";print ""
##################

#####

print "..tuned on training set only...";print ""

clf_t = svm.SVC(kernel='poly',C=best_C,degree=best_degree)
clf_t.fit(X_train0, y_train0)
print "Training Set Accuracy:", clf_t.score(X_train0,y_train0)
print "Validation Set Accuracy:", clf_t.score(X_valid0,y_valid0)
print "Test Set Accuracy:", clf_t.score(X_test,y_test)

print ""

#####

print "..tuned on TOTAL training set (train together with validation)...";print ""

clf_t2 = svm.SVC(kernel='poly',C=best_C,degree=best_degree)
clf_t2.fit(X_train, y_train)
print "Training Set Accuracy (Total Training Set):", clf_t2.score(X_train,y_train)
print "Test Set Accuracy:", clf_t2.score(X_test,y_test)

trial_params = np.array([best_degree,best_C,current_best_acc])
```

#### Try without folds.  Just straight testing for each set of params:

```python
# Tune C and Degree via cross-validation

# no folding.  straight validation for each parameter try.

Optimal_C = []

for deg in range(1,5,1):

    #C = np.exp(np.linspace(-10,2*i,200))
    C = [math.exp(i) for i in np.linspace(-10,2*deg,200)] ## get in list form instead, no real reason
    OS_validation=[]

    # iterate through Cs
    for c in C:
        clf = svm.SVC(kernel='poly',C=c,degree=deg) 
        clf.fit(X_train0, y_train0)
        acc_per_C = 1.0*clf.score(X_valid0, y_valid0) # which validation set are we actually using here (??)
        OS_validation.append(acc_per_C)
        
    # record optimal C for given degree
    
    ### novice weird code 
    
    temp=pd.DataFrame([C,OS_validation]).T
    mid_C=len(temp.loc[temp.iloc[:,1]==temp.iloc[:,1].max()])/2
    C_opt=temp.loc[temp.iloc[:,1]==temp.iloc[:,1].max()].iloc[mid_C,0]
    
    # let's check out accuracy distributions across hyperparams
    
    C_plot=[math.log(y,10) for y in C] # visualize (at log scale)
    
    pylab.plot(C_plot,OS_validation,'b',)
    pylab.legend(loc='upper right')
    pylab.ylabel('Accuracy')
    pylab.xlabel('log(C)')
    pylab.show()
        
    print "For Degree ",deg; print "Optimal C :",C_opt,"@ max accuracy (",temp.iloc[:,1].max(),")";print"";print""
          
    Optimal_C.append([deg,C_opt,temp.iloc[:,1].max()]) # record best for each degree
```

```python
# report best hyperparameters from cross-valid

OPT = pd.DataFrame(Optimal_C)
parms = OPT.loc[OPT.iloc[:,2] == OPT.iloc[:,2].max()].values[0]
acc_max, degree_best, c_best = parms[2],parms[0],parms[1]


print "\nAccuracy is at its max (",acc_max,") when degree is",degree_best,"and C is", c_best

print "";print "---"


################

#####

print "..tuned on training set only...";print ""

clf_t = svm.SVC(kernel='poly',C=c_best,degree=degree_best)
clf_t.fit(X_train0, y_train0)
print "Training Set Accuracy:", clf_t.score(X_train0,y_train0)
print "Validation Set Accuracy:", clf_t.score(X_valid0,y_valid0)
print "Test Set Accuracy:", clf_t.score(X_test,y_test)

print ""

#####

print "..tuned on TOTAL training set (train together with validation)...";print ""

clf_t2 = svm.SVC(kernel='poly',C=c_best,degree=degree_best)
clf_t2.fit(X_train, y_train)
print "Training Set Accuracy (Total Training Set):", clf_t2.score(X_train,y_train)
print "Test Set Accuracy:", clf_t2.score(X_test,y_test)

trial_params = np.append(trial_params,np.array([degree_best,c_best,acc_max])).reshape(2,3)
```
